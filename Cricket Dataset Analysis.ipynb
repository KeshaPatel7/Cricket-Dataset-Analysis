```python
odi_df['Host_Country'] = odi_df['Host_Country'].fillna('Neutral')
odi_df.info()
```

# Dropping rows/matches that was played for two days

```python
odi_df.drop(odi_df[odi_df['Match Date'].str.contains('-')].index, inplace = True)
odi_df.info()
```

# Adding new column "Match_Day" derived from "Match Date" as datetime datatype to extract matches based on date

```python
odi_df['Match_Day'] = odi_df['Match Date'].str.replace(' ', '')
odi_df['Match_Day'] = pd.to_datetime(odi_df['Match_Day'], format='%b%d,%Y')
odi_df.head()
```

# Considering data of this century(2000 & above) alone to see insights of modern times

```python
odi_df = odi_df[odi_df['Match_Day'] >= "2000-01-01"]
odi_df.head()
```

# Removing matches that are populated twice in account of Team perspective(Only Unique ODI numbers are considered) by duplicated indices

```python
odi_df.sort_values(by = ['Scorecard', 'Venue_Team1'], inplace = True, ascending = [True, False])
odi_df.head()
```

```python
odi_df = odi_df[~odi_df.index.duplicated(keep='first')]
odi_df
```

# Visualization

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
winning_team = odi_df['Winner'].value_counts().index[:10]
win_count = odi_df['Winner'].value_counts().values[:10]

plt.figure(figsize=(12,8))
ax = sns.barplot(y = winning_team , x = win_count, orient='h', palette='winter')
ax.set_xlabel(xlabel='Team Names', fontdict = {'size' : 'x-large', 'fontweight' : 'bold'})
ax.set_ylabel(ylabel='Wins', fontdict = {'size' : 'x-large', 'fontweight' : 'bold'})
ax.set_title("Top 10 Teams by Wins", fontdict = {'size' : 'xx-large', 'fontweight' : 'bold'})
ax.bar_label(ax.containers[0])
ax.set_yticklabels(winning_team, fontsize=14)
plt.show()
```

Australia has won most numbers of matches. 

```python
team1_by_mat = odi_df.groupby('Team 1').agg({'Scorecard':'count'}).reset_index()
team2_by_mat = odi_df.groupby('Team 2').agg({'Scorecard':'count'}).reset_index()

team1_by_mat.rename(columns={"Team 1": "idx_team", "Scorecard": "Matches"}, inplace = True)
team2_by_mat.rename(columns={"Team 2": "idx_team", "Scorecard": "Matches"}, inplace = True)

total_mat = pd.concat([team1_by_mat, team2_by_mat]).groupby(['idx_team']).sum().reset_index()
total_mat
```

```python
plt.figure(figsize=(22,8))

ax = sns.barplot(y = total_mat['Matches'] , x = total_mat['idx_team'], palette='plasma')
ax.set_xlabel(xlabel='Teams', fontdict = {'size' : 'x-large', 'fontweight' : 'bold'})
ax.set_ylabel(ylabel='Number of Matches', fontdict = {'size' : 'x-large', 'fontweight' : 'bold'})
ax.set_title("Matches Played By Team", fontdict = {'size' : 'xx-large', 'fontweight' : 'bold'})
ax.bar_label(ax.containers[0])
plt.show()
```

India has played most number of matches

```python
colors=["gold","green","lightsteelblue","firebrick","royalblue"]
country = odi_df['Host_Country'].value_counts().head().index
mat_count = odi_df['Host_Country'].value_counts().head().values

plt.figure(figsize=(10, 8))
plt.title("Top 5 Countries by Matches Hosted", fontdict = {'size' : 'xx-large', 'fontweight' : 'bold'})
plt.pie(mat_count, labels=country, autopct='%0.2f%%',
        colors = colors, startangle=60, explode = [0.06,0.02,0.02,0.02,0.02], 
        textprops={'color':"black", 'size' : 'x-large', 'fontweight' : 'bold'})
plt.show()
```
